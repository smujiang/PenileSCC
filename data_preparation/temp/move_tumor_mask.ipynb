```python
import os
import shutil
```

```python
data_root_dir = "/projects/wangc/jun/data/Penile_SCC_Study/processed_annotations"

case_id_list = os.listdir(data_root_dir)


for case_id in case_id_list:
    print(case_id)
    img_fn_list = os.listdir(os.path.join(data_root_dir, case_id))
    for img_fn in img_fn_list:
        if "Thumb" in img_fn:
            os.remove(os.path.join(data_root_dir, case_id, img_fn))
        if "Tumor-mask" in img_fn:
            tumor_mask_fn = os.path.join(data_root_dir, case_id, img_fn)
            new_fn = img_fn.replace("Tumor-mask_", "").replace(".png", "-mask.png")
            rename_to = os.path.join(data_root_dir, case_id, new_fn)
            print(tumor_mask_fn)
            print(rename_to)
            shutil.move(tumor_mask_fn, rename_to)
            
```
